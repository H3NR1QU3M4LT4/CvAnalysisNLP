# AmilcarCarvalho

```python
dataframe = pd.read_csv('drive/MyDrive/data_analysis/data/csv_data_analysis/AmilcarCarvalho-EN_analysis.csv')
```

```python
dataframe
```

```python
label_name_label_studio = {
    0: "NONE",
    1: "SUMMARY",
    2: "EDUCATION",
    3: "WORK_EXPERIENCE",
    4: "PERSONAL_DETAILS",
    5: "TECHNICAL_SKILLS",
    6: "LANGUAGES",
    7: "OTHER",
}
```

```python
y = dataframe['labels']
```

```python
X = dataframe.drop("text", axis='columns')
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=8)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

centers = kmeans.cluster_centers_
```

```python
import numpy as np
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)

principalDf = pd.DataFrame(data = principalComponents, columns = ['x', 'y'])
```

```python
y = y.to_numpy()
```

```python
plotty_data = principalDf
plotty_data['kmean_labels'] = y_kmeans
plotty_data['true_labels'] = y

plotty_data['kmean_labels_string'] = plotty_data['kmean_labels'].map(label_name_label_studio)
plotty_data['true_labels_string'] = plotty_data['true_labels'].map(label_name_label_studio)
```

```python
plotty_data
```

```python
fig = px.scatter(plotty_data, x="x", y="y", color="kmean_labels")
fig.update_layout(showlegend=False)
fig.show()
```

```python
fig = px.scatter(plotty_data, x="x", y="y", color="true_labels_string", 
                 hover_data=['true_labels_string'], 
                 symbol="true_labels_string", 
                 title="True Labels")
fig.show()
```

```python
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score

# Compute the ARI and NMI scores
ari_score = adjusted_rand_score(plotty_data['true_labels'], plotty_data['kmean_labels'])
nmi_score = normalized_mutual_info_score(plotty_data['true_labels'], plotty_data['kmean_labels'])

# Print the scores
print(f"ARI score: {ari_score}")
print(f"NMI score: {nmi_score}")
```

# ARI and NMI for all

```python
import pandas as pd
import os
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score

label_name_label_studio = {
    0: "NONE",
    1: "SUMMARY",
    2: "EDUCATION",
    3: "WORK_EXPERIENCE",
    4: "PERSONAL_DETAILS",
    5: "TECHNICAL_SKILLS",
    6: "LANGUAGES",
    7: "OTHER",
}

ari_scores = []
nmi_scores = []

for filename in os.listdir("drive/MyDrive/data_analysis/data/csv_data_analysis/"):
    df = pd.read_csv(f"drive/MyDrive/data_analysis/data/csv_data_analysis/{filename}")
    
    # Extract labels and features
    y = df['labels']
    X = df.drop("text", axis='columns')
    
    # Apply clustering algorithm
    kmeans = KMeans(n_clusters=8)
    kmeans.fit(X)
    y_kmeans = kmeans.predict(X)
    
    # Compute performance metrics
    ari_score = adjusted_rand_score(y, y_kmeans)
    nmi_score = normalized_mutual_info_score(y, y_kmeans)
    
    # Append scores to lists
    ari_scores.append(ari_score)
    nmi_scores.append(nmi_score)
```

```python
b=np.sum(nmi_scores) 
b
```

```python
# Compute the overall ARI and NMI scores
overall_ari = np.sum(ari_scores) / len(ari_scores)
overall_nmi = np.sum(nmi_scores) / len(nmi_scores)

# Print the scores
print(f"Overall ARI score: {overall_ari}")
print(f"Overall NMI score: {overall_nmi}")
```
